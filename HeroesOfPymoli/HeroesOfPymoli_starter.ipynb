```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#usernames without duplicates #df same as dataframe lectures
jugadores = len(purchase_data["SN"].unique())

players_df = {'Total Players': [jugadores]}
jugadores_df = pd.DataFrame(players_df, columns=["Total Players"])
jugadores_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculations as data functions
items = len(purchase_data["Item ID"].unique())
average = purchase_data["Price"].mean()
purchases = len(purchase_data["Purchase ID"].unique())
revenue = purchase_data["Price"].sum()

resumen_df = {'Number of Unique Items': [items], 'Average Price': '${:.2f}'.format(average), 'Number of Purchases': [purchases], 'Total Revenue': '${:.2f}'.format(revenue)}
analysis_df = pd.DataFrame(resumen_df, columns=["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"])
analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#dup out
no_dup_df = purchase_data.drop_duplicates(subset='SN', keep="last")
genderdata_df = no_dup_df[['SN','Gender']]

male = len(genderdata_df.loc[genderdata_df["Gender"] == "Male",:])
female = len(genderdata_df.loc[genderdata_df["Gender"] == "Female",:])
other = len(genderdata_df.loc[genderdata_df["Gender"] == "Other / Non-Disclosed",:])

addcount = genderdata_df["Gender"].count()
mper = (male/addcount)*100
fper = (female/addcount)*100
oper = (other/addcount)*100


sum_demo = pd.DataFrame({'Total Count':[male, female, other], 'Percentage of Players':[mper, fper, oper]})
sum_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#no dub/remove uneeded
gender_df = purchase_data[['SN','Gender','Price']]
gen_nodub = gender_df.drop_duplicates(subset='SN', keep="last")

purchase = gender_df.groupby(["Gender"])["Price"].count()
item= gender_df.groupby(["Gender"])["Price"].mean()
total_value = gender_df.groupby(["Gender"])["Price"].sum()
solo = gen_nodub.groupby(["Gender"])["Price"].count()
ave_total =(total_value/solo)

summary_gender = pd.DataFrame({"Purchase Count":purchase, "Item Price": round((item),2), "Total Purchase Value":(total_value),"Avg Total Purchase per Person":round(ave_total,2)})
summary_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#as binning lecture
no_dup_df = purchase_data.drop_duplicates(subset='SN', keep="last")
bin = [0, 9,  14, 19, 24, 29, 34, 39, 100]
label = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#no dup
copy = no_dup_df.copy()
copy["Groups"] = pd.cut(copy["Age"], bin, labels=label)
age= copy.groupby(["Groups"])

age_count = age["SN"].count()
total_count = copy["SN"].count()
percentage = (age_count / total_count)*100

age_demo_df = pd.DataFrame({"Total Count": age_count,"Percentage of Players": round(percentage,2)})
age_demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bin = [0, 9, 14, 19, 24, 29, 34, 39, 150]
label = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

copy = purchase_data.copy()
copy["Age Groups"] = pd.cut(copy["Age"], bin, labels=label)
age = copy.groupby(["Age Groups"])

age_df= age["SN"].count()
total = copy["SN"].count()
percentage = (age_df/ total)*100
mean = age["Price"].mean()
suma = age["Price"].sum()
percentagetotal = (total/age_df)

age_dem = pd.DataFrame({"Purchase Count": age_df,"Average Purchase Price":round(mean,2),"Total Purchase Value":round(suma,2), "Avg Total Purchase per Person":round(percentagetotal,2)})
age_dem
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#drop columns
sn_df = purchase_data[['SN', 'Price']]

purchase = sn_df.groupby(["SN"])["Price"].count()
price = sn_df.groupby(["SN"])["Price"].mean()
total = sn_df.groupby(["SN"])["Price"].sum()

spenders = pd.DataFrame({"Purchase Count":purchase, "Average Purchase Price":round(price,2), "Total Purchase Value":total})
sort = spenders.sort_values("Purchase Count", ascending= False)
sort.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#drop columns
popular_df = purchase_data[['Item ID', 'Item Name', 'Price']]

#calculate variables for table
items = popular_df.groupby(["Item ID","Item Name"])["Price"].count()
ave= popular_df.groupby(["Item ID","Item Name"])["Price"].mean()
total = popular_df.groupby(["Item ID","Item Name"])["Price"].sum()

#obtain new df and sort. Print the top 5 results
popular_items = pd.DataFrame({"Purchase Count":items, "Item Price":ave, "Total Purchase Value":total})
sort = popular_items.sort_values("Purchase Count", ascending= False)
sort.head()
```
